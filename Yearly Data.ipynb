```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bat = pd.read_csv('batting.csv')
bowl = pd.read_csv('bowling.csv')
```

```python
bat.head()
```

```python
b = bat
b.head()
```

```python
b.info()
```

```python
b['4s'] = pd.to_numeric(b['4s'],errors = 'coerce')
b['6s'] = pd.to_numeric(b['6s'],errors = 'coerce')
b['SR'] = pd.to_numeric(b['SR'],errors = 'coerce')
b.info()
```

```python
b = pd.DataFrame(b.groupby(['Name', 'Year']).agg({'Runs': 'sum', '4s': 'sum', '6s': 'sum', 'SR': 'mean'}).reset_index())
```

```python
b
```

```python
b.to_csv('batting_consolidated.csv',index=False)
```
