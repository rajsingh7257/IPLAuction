```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, RandomForestClassifier, GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import metrics
```

## Pre-processing for Model

```python
bat_us = batting.drop(['Name','PID','PRICE'],axis=1)
bat_us
```

```python
bat_us['NATIONALITY'] = np.where(bat_us['NATIONALITY']=='Overseas',1,0)
bat_us['SOLD/UNSOLD'] = np.where(bat_us['SOLD/UNSOLD']=='SOLD',1,0)
bat_us
```

```python
bat_us.info()
```

```python
X_bat_us = bat_us.drop(['SOLD/UNSOLD'],axis=1)
y_bat_us = bat_us['SOLD/UNSOLD']
```

```python
X_train_bat_us, X_test_bat_us, y_train_bat_us, y_test_bat_us = train_test_split(X_bat_us,y_bat_us,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
X_train_bat_us = minmax.fit_transform(X_train_bat_us)
X_test_bat_us = minmax.transform(X_test_bat_us)
```

### Logistic

```python
model = LogisticRegression()
model.fit(X_train_bat_us,y_train_bat_us)
```

```python
y_train_predict = model.predict(X_train_bat_us)
model_score = model.score(X_train_bat_us,y_train_bat_us)
print(model_score)
print(metrics.confusion_matrix(y_train_bat_us,y_train_predict))
print(metrics.classification_report(y_train_bat_us,y_train_predict))
```

```python
y_test_predict = model.predict(X_test_bat_us)
model_score = model.score(X_test_bat_us,y_test_bat_us)
print(model_score)
print(metrics.confusion_matrix(y_test_bat_us,y_test_predict))
print(metrics.classification_report(y_test_bat_us,y_test_predict))
```

### LDA

```python
model = LinearDiscriminantAnalysis()
model.fit(X_train_bat_us,y_train_bat_us)
```

```python
y_train_predict = model.predict(X_train_bat_us)
model_score = model.score(X_train_bat_us,y_train_bat_us)
print(model_score)
print(metrics.confusion_matrix(y_train_bat_us,y_train_predict))
print(metrics.classification_report(y_train_bat_us,y_train_predict))
```

```python
y_test_predict = model.predict(X_test_bat_us)
model_score = model.score(X_test_bat_us,y_test_bat_us)
print(model_score)
print(metrics.confusion_matrix(y_test_bat_us,y_test_predict))
print(metrics.classification_report(y_test_bat_us,y_test_predict))
```
